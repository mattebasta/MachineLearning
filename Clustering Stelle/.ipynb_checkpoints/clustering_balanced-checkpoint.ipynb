```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
import re
import seaborn as sns
from sklearn import svm

data = pd.read_csv("Star39552_balanced.csv", sep=",")
data.head()
```

```python
replace= {'K([0-9a-zA-Z].*)':'K','F([0-9a-zA-Z].*)':'F','B([0-9a-zA-Z].*)':'B','O([0-9a-zA-Z].*)':'O','A([0-9a-zA-Z].*)':'A','G([0-9a-zA-Z].*)':'G','M([0-9a-zA-Z].*)':'M','C([0-9a-zA-Z].*)':'C'}
data_new= data.replace({"SpType": replace}, regex=True)
```

```python
data_new
```

```python
NType = {'O': 0, 'B': 1, 'A': 2, 'F': 3, 'G': 4, 'K': 5, 'M': 6, 'C': 7}
data_new['NumSpType'] = data_new['SpType'].map(NType)
data_new
```

```python
Sum_Type = data_new["NumSpType"].value_counts()
Sum_Type
```

```python
Sum_Type_list =list(Sum_Type)
Sum_Type_list
```

```python
data_new.dropna(subset = ["NumSpType"], inplace=True)
```

```python
X = data_new[['Vmag','Plx','B-V','Amag']]
y = data_new.NumSpType
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth=10, random_state=42)
rfc.fit(X,y)
```

```python
print(rfc.feature_importances_)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
select = SelectKBest(f_classif, k=2)
select.fit(X, y)
```

```python
new_X = data_new[['B-V','Amag']]
new_Y = data_new.NumSpType
df_new = pd.DataFrame(new_X)
df_new
```

```python
new_Y.value_counts()
```

```python
f, ax2 = plt.subplots(figsize=(18,12))
sns.scatterplot(x=df_new.iloc[:, 0], y=df_new.iloc[:, 1], s=60, c=new_Y, legend="full")
```

```python
f, ax1 = plt.subplots(figsize=(12,10))
sns.scatterplot(x=df_new.iloc[:, 0], y=df_new.iloc[:, 1], s=5, color=".15")
sns.histplot(x=df_new.iloc[:, 0], y=df_new.iloc[:, 1], bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=df_new.iloc[:, 0], y=df_new.iloc[:, 1], levels=5, color="w", linewidths=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_X, new_Y, test_size=0.40, random_state=42)
#-------------------------RANDOMFOREST-----------------------------RANDOMFOREST-----------------------------------
print(rfc.fit(X_train, y_train))
```

```python
print("Train Set",rfc.score(X_train,y_train))
print("Test Set",rfc.score(X_test,y_test), "\n")
print("Train set lenght: ",len(X_train))
print("Test set lenght: ",len(X_test))
```

```python
predict=rfc.predict(X_test)
```

```python
cm= confusion_matrix(y_test, predict)
print("CM test set\n", cm)
```

```python
df_cm = pd.DataFrame(cm)
plt.figure(figsize=(12, 8))
sns.set(font_scale=1)
sns.heatmap(df_cm, annot=True, fmt="d", annot_kws={"size":20})
plt.show()
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, predict)
print('Accuracy: %.3f' % acc)
```

```python
#----------------------------------------SVM-----------------------SVM--------------------------------------------
csvm = svm.SVC(kernel="rbf", gamma=0.7, C=1.0)
print(csvm.fit(X_train, y_train))
```

```python
print("Train Set",csvm.score(X_train,y_train))
print("Test Set",csvm.score(X_test,y_test), "\n")
```

```python
svm_predict=csvm.predict(X_test)
cm1=confusion_matrix(y_test, svm_predict)
print("CM SVM\n",cm1)
```

```python
df_cm1 = pd.DataFrame(cm1)
plt.figure(figsize=(12, 8))
sns.set(font_scale=1)
sns.heatmap(df_cm1, annot=True, fmt="d", annot_kws={"size":20})
plt.show()
```

```python
acc1=accuracy_score(y_test, svm_predict)
print('Accuracy: %.3f' % acc1)
```

```python
#---------------------------------KNN----------------------------------KNN----------------------------------------
from sklearn import neighbors
error1=[]
error2=[]
for k in range(1,20):
    knn=neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred1= knn.predict(X_train)
    error1.append(np.mean(y_train!=y_pred1))
    y_pred2= knn.predict(X_test)
    error2.append(np.mean(y_test!=y_pred2))

plt.figure(figsize=(12,10))
plt.plot(range(1,20), error1, label="train_Set")
plt.plot(range(1,20), error2, label="test_Set")
plt.xlabel('K value')
plt.ylabel('Error')
plt.legend()
```

```python
knn=neighbors.KNeighborsClassifier(n_neighbors=20)
print(knn.fit(X_train, y_train))
```

```python
print("Train Set",knn.score(X_train,y_train))
print("Test Set",knn.score(X_test,y_test), "\n")
```

```python
knn_predict=knn.predict(X_test)
cm2=confusion_matrix(y_test, knn_predict)
print("CM SVM\n",cm2)
```

```python
df_cm2 = pd.DataFrame(cm2)
plt.figure(figsize=(12, 8))
sns.set(font_scale=1)
sns.heatmap(df_cm2, annot=True, fmt="d", annot_kws={"size":20})
plt.show()
```

```python
acc2=accuracy_score(y_test, knn_predict)
print('Accuracy: %.3f' % acc2)
```
