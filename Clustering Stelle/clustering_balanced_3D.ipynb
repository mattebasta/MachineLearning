```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import re

data = pd.read_csv("Star39552_balanced.csv", sep=",")
data.head()
```

```python
data
```

```python
print(data.dtypes)
```

```python
replace= {'K([0-9a-zA-Z].*)':'K','F([0-9a-zA-Z].*)':'F','B([0-9a-zA-Z].*)':'B','O([0-9a-zA-Z].*)':'O','A([0-9a-zA-Z].*)':'A','G([0-9a-zA-Z].*)':'G','M([0-9a-zA-Z].*)':'M','C([0-9a-zA-Z].*)':'C'}
data_new= data.replace({"SpType": replace}, regex=True)
```

```python
data_new
```

```python
NType = {'K': 0, 'F': 1, 'B': 2, 'O': 3, 'A': 4, 'G': 5, 'M': 6, 'C': 7}
data_new['NumSpType'] = data_new['SpType'].map(NType)
data_new
```

```python
Sum_Type = data_new["NumSpType"].value_counts()
Sum_Type
```

```python
data_new.dropna(subset = ["NumSpType"], inplace=True)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
X = data_new[['Vmag','Plx','B-V','Amag']]
y = data_new.NumSpType
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
print(clf.fit(X_train, y_train))
```

```python
clf.feature_importances_
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
select = SelectKBest(f_classif, k=3)
select.fit(X, y)
```

```python
mask = select.get_support()
print(mask)
```

```python
X_3D = X.iloc[:, [True, False, True, True]]
X_3D
```

```python
df_3D = pd.DataFrame(X_3D)
df_3D.head()
```

```python
kmeans = KMeans(n_clusters=8,random_state=42).fit(df_3D)
clusters = kmeans.cluster_centers_
clusters
```

```python
fig=plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df_3D.iloc[:, 0], df_3D.iloc[:, 1], df_3D.iloc[:, 2])
ax.set_xlabel('Vmag')
ax.set_ylabel('B-V')
ax.set_zlabel('Amag')
plt.show()
```

```python
X_3D_arr = df_3D.to_numpy()
X_3D_arr
```

```python
Y_arr = y.to_numpy()
Y_arr
```

```python
Y_arr = Y_arr.astype(int)
Y_arr
```

```python
pairs = {}
predicted = []
for i in range(len(X_3D_arr)):
    predicted_class = kmeans.predict([X_3D_arr[i]])[0]
    original_class = Y_arr[i]
   #print([X[i]], predicted_class, original_class)
    predicted.append((X_3D_arr[i], predicted_class))
    pair = (predicted_class, original_class)
    pairs[pair] = pairs.get(pair, 0)
    pairs[pair] += 1
    

for pair in pairs:
    print(pair, pairs[pair])
```
